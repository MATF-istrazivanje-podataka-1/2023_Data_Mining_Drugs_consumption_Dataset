```python
podaci_opisno = podaci.copy()
```

```python

godine = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
pol = ['M', 'Z']
obrazovanje = ['N < 16', 'N 16', 'N 17', 'N 18', 'N Fakultet', 'Zanat', 'Fakultet', 'Master', 'Doktorat']
drzavljanstvo = ['USA', 'NZ', 'Ostalo', 'AUS', 'IR', 'CAN', 'UK']
etnicko_poreklo = ['Negroidno', 'Mongolidno', 'Evropeidno', 'Mulat', 'Ostalo', 'Melez', 'Zambos']


# print(len(etnicko_poreklo), len(vrednosti_atributa['Ethnicity']))

def opisi_podatke(atribut, niz, opisni_niz):
    n = len(niz)
    mapa_podataka = {}
    for i in range(n):
        mapa_podataka.update({niz[i]: opisni_niz[i]})
        #print(f'{niz[i]}:{opisni_niz[i]}')
    podaci_opisno[atribut] = podaci_opisno[atribut].map(mapa_podataka)
    
opisi_podatke('Age', vrednosti_atributa['Age'], godine)
opisi_podatke('Gender', vrednosti_atributa['Gender'], pol)
opisi_podatke('Education', vrednosti_atributa['Education'], obrazovanje)
opisi_podatke('Country', vrednosti_atributa['Country'], drzavljanstvo)
opisi_podatke('Ethnicity', vrednosti_atributa['Ethnicity'], etnicko_poreklo)

podaci_opisno
```

```python

# fja standardizacije (minmax)
def skaliraj_podatke(X):
    return (X - min(X))/(max(X) - min(X))
```

```python
skalirani_podaci = podaci.copy()
for a in atributi:
    skalirani_podaci[a] = skaliraj_podatke(skalirani_podaci[a])
    
#provera uspesnosti skaliranja
for a in atributi:
    print(len(skalirani_podaci[a].unique()) == len(podaci[a].unique()), end = ", ")
    
#nismo izgubili podatke
```

Stepen korelacije medju podacima

```python
podaci_svi = podaci.copy()

def drug_encoder(x):
    if x == 'CL0':
        return 0
    elif x == 'CL1':
        return 1
    elif x == 'CL2':
        return 2
    elif x == 'CL3':
        return 3
    elif x == 'CL4':
        return 4
    elif x == 'CL4':
        return 5
    elif x == 'CL5':
        return 6
    else:
        return 7
```

```python
for klasa in klase:
    podaci_svi[klasa] = podaci[klasa].apply(drug_encoder)
```

```python
corr = podaci_svi.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr, annot=True, vmin=-1)
```
