```python
# raspon vrednosti atributa

vrednosti_atributa = {}
broj_vrednosti_atributa = {}
for atribut in podaci.columns:
    vrednosti_atributa[atribut] = podaci[atribut].unique()
    broj_vrednosti[atribut] = len(podaci[atribut].unique())

print(f'Broj vrednosti po atributu:\n {broj_vrednosti}\n')
# print(f'Vrednosti po atributu: \n {vrednosti_atributa}\n')
vrednosti_atributa
```

## PREPROCESIRANJE

```python
#nedostajuce vrednosti
podaci.isna().any().any()
```

```python
podaci = podaci.drop('ID', axis = 1) # ne treba nam atribut ID, ne govori nam nista
```

```python
# menjam realne const. u adekvatne vrednosti za kategoricke podatke
#sortiram moguce vrednosti prvo
for atribut in vrednosti_atributa:
    vrednosti_atributa[atribut].sort() 
```

```python
podaci_opisno = podaci.copy()
```

```python

godine = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
pol = ['M', 'Z']
obrazovanje = ['N < 16', 'N 16', 'N 17', 'N 18', 'N Fakultet', 'Zanat', 'Fakultet', 'Master', 'Doktorat']
drzavljanstvo = ['AUS', 'CAN', 'NZ', 'Ostalo', 'IR', 'UK', 'USA']
etnicko_poreklo = ['Mongoloidno', 'Negroidno', 'Zambos', 'Melez', 'Mulat', 'Ostalo', 'Evropeidno']


print(len(etnicko_poreklo), len(vrednosti_atributa['Ethnicity']))

def opisi_podatke(atribut, niz, opisni_niz):
    n = len(niz)
    mapa_podataka = {}
    for i in range(n):
        mapa_podataka.update({niz[i]: opisni_niz[i]})
        #print(f'{niz[i]}:{opisni_niz[i]}')
    podaci_opisno[atribut] = podaci_opisno[atribut].map(mapa_podataka)
    
opisi_podatke('Age', vrednosti_atributa['Age'], godine)
opisi_podatke('Gender', vrednosti_atributa['Gender'], pol)
opisi_podatke('Education', vrednosti_atributa['Education'], obrazovanje)
opisi_podatke('Country', vrednosti_atributa['Country'], drzavljanstvo)
opisi_podatke('Ethnicity', vrednosti_atributa['Ethnicity'], etnicko_poreklo)

podaci_opisno
```
