## ocenjivanje modela

```python
from sklearn.metrics import silhouette_samples, silhouette_score

def crtaj_grafik(ocene, ocena):
    plt.figure(figsize=(8, 4))
    plt.bar(imena_modela, ocene, color=['blue', 'red', 'green'])
    plt.grid(alpha=0.5)
    for i, s in enumerate(ocene):
        plt.text(imena_modela[i], s, f'{s:.4f}', ha='center', va='bottom')
    plt.title(f'{ocena} ocena za modele')
    plt.show()
```

```python
crtaj_grafik(senka_ocene, "Senka")
```

### Dunn indeks

```python
from scipy.spatial.distance import pdist, squareform

def dunn_index(X, labels):
    inter_cluster_dist = np.min(squareform(pdist(X))[labels != labels[:, np.newaxis]])
    intra_cluster_dist = np.max([np.max(pdist(X[labels == i])) for i in np.unique(labels)])
    dunn_index = inter_cluster_dist / intra_cluster_dist
    
    return dunn_index
```

```python
dunn_ocene = []
for model in modeli:
    dunn_ocene.append(dunn_index(X, model.labels_))
```

```python
dunn_ocene
```

```python
crtaj_grafik(dunn_ocene, "Dunn")
```

### Intra klaster razlike

```python
from sklearn.metrics import pairwise_distances

def calculate_intra_cluster_distance(X, labels):
    distances = pairwise_distances(X)
    intra_cluster_distance = 0.0
  
    unique_labels = set(labels)
    for label in unique_labels:
        indices = (labels == label)
        intra_cluster_distance += np.sum(distances[np.ix_(indices, indices)])
    
    return intra_cluster_distance
```

```python
intraRastojanja_ocene = []
for model in modeli:
      intraRastojanja_ocene.append(calculate_intra_cluster_distance(X, model.labels_))

intraRastojanja_ocene
```

```python
crtaj_grafik(intraRastojanja_ocene, "Intra rastojanja")
```

### Balansiranost klasa

```python
fig = plt.figure(figsize=(16, 4))
plt.suptitle('Balansiranost predvidjenih klastera')
idx = 1
for i, model in enumerate(modeli):
  
    pieData = pd.Series(model.labels_).value_counts()
    pieData.reset_index(drop=True, inplace=True)
    p = fig.add_subplot(1, 4, idx)
    idx += 1
    p.set_title(imena_modela[i])

    plt.pie(pieData, startangle= 90, autopct= '%.1f%%', pctdistance=1.2)

plt.show()
```
