```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#uzeto iz fajla za analizu podataka
kolone = ["ID", "Age", "Gender", "Education", "Country", "Ethnicity", 
            "Nscore", "Escore", "Oscore", "Ascore", "Cscore", "Impulsiveness", "SS", 
            "Alchocol", "Amphet", "Amyl", "Benzos", "Caff", "Canabis", "Choc", "Coke", "Crack", "Ecstasy", 
            "Heroin", "Ketamin", "LegalH", "LSD", "Meth", "Mushrooms", "Nicotine", "Semer", "VSA"]

podaci = pd.read_csv('drug_consumption.data', names = kolone)

podaci = podaci.drop('ID', axis = 1)
podaci 
```

```python
klase = []
atributi = []
for x in podaci.columns:
    if x == 'Alchocol':
        break
    atributi.append(x)
print(atributi)

for x in podaci.columns:
    if x not in atributi:
        klase.append(x)
print(klase)
```

```python
#podesavanje podataka malo
for x in klase:
    a = podaci[x].unique()
    a.sort()
    print(a)
```

```python
podaci_skalirano = podaci.copy()

opisni_niz1 = podaci['Alchocol'].unique()
opisni_niz1.sort()
n = len(podaci_skalirano['Heroin'])
    
for x in klase:
    for i in range(n):
        podaci_skalirano[x][i] = int(podaci_skalirano[x][i].strip('CL'))
    print(podaci_skalirano[x].unique())

podaci_skalirano
#podaci_skalirano['Alchocol'].unique()
```

```python
for x in klase:
    podaci_skalirano[x] = podaci_skalirano[x].astype('float64')
    
podaci_skalirano.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
klasa = podaci_skalirano['Alchocol']
podaci_bez_klase = podaci_skalirano.drop('Alchocol', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(podaci_bez_klase, klasa, test_size= 0.30, stratify= klasa, random_state= 42)
```

```python
drvo_klasifikacije = DecisionTreeClassifier()
drvo_klasifikacije.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
from termcolor import colored, cprint
def report(model, x, y, text = "trening", ):
    y_pred = model.predict(x)
    
    print(colored('\t' + f'Klasifikacioni izveštaj za model {type(model).__name__} nad {text} podacima', 'blue', attrs= ['bold']))
    cprint('----------------------------------------------------------------------------------------------', attrs= ['bold'])
    cprint(classification_report(y, y_pred), attrs= ['dark']) 
    cprint('----------------------------------------------------------------------------------------------', attrs= ['bold'])
```

```python
report(drvo_klasifikacije, X_test, y_test, text= 'test')
```

```python
pd.Series(drvo_klasifikacije.feature_importances_, index=podaci_bez_klase.columns).plot.barh()
plt.title("Feature importance")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
parametri =  {
    'criterion': ['gini', 'entropy'],
    'max_depth': [x for x in range(3, 30)],
    'max_features': [x for x in range(3, 40)]
}

balansirano_drvo = GridSearchCV(DecisionTreeClassifier(class_weight= 'balanced'), param_grid= parametri)
```

```python
balansirano_drvo.fit(X_train, y_train)
```

```python
report(balansirano_drvo, X_test, y_test)
```

```python
print(balansirano_drvo.best_params_)
print(balansirano_drvo.best_score_)
```
